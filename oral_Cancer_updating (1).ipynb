````python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
import matplotlib.pyplot as plt


IMG_SIZE = (224, 224)
BATCH_SIZE = 32
EPOCHS = 60
LEARNING_RATE = 0.0001
DATASET_PATH = '/content/train'


def preprocess_image(image, label):
    image = tf.image.resize(image, IMG_SIZE) / 255.0
    return image, label

def load_dataset(dataset_path, batch_size, subset):
    dataset = tf.keras.utils.image_dataset_from_directory(
        dataset_path,
        label_mode="binary",
        image_size=IMG_SIZE,
        batch_size=batch_size,
        validation_split=0.2,
        subset=subset,
        seed=123
    )
    dataset = dataset.map(preprocess_image).prefetch(buffer_size=tf.data.AUTOTUNE)
    return dataset

train_dataset = load_dataset(DATASET_PATH, BATCH_SIZE, subset="training")
val_dataset = load_dataset(DATASET_PATH, BATCH_SIZE, subset="validation")

all_labels = []
for _, labels in train_dataset.unbatch():
    all_labels.append(labels.numpy())
all_labels = np.concatenate(all_labels, axis=0)

class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(all_labels),
    y=all_labels
)
class_weights = dict(enumerate(class_weights))
print("Class Weights:", class_weights)



base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
              loss='binary_crossentropy',
              metrics=['accuracy'])

callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1),
    ModelCheckpoint('Oral_Cancer_Model.keras', save_best_only=True, monitor='val_loss', verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1, min_lr=1e-6)
]


model.summary()


history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=EPOCHS,
    class_weight=class_weights,
    callbacks=callbacks
)


model.save('Oral_Cancer_Model.keras')

plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy Over Epochs')
plt.show()

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss Over Epochs')
plt.show()
````

# Task
The current model, while effective in classifying biopsy images, is a binary classifier trained on 'Normal' and 'OSCC' biopsy slides. This means that if it encounters an image that is *not* a biopsy slide at all (an "irrelevant" image, such as a photo of a document, a landscape, or a completely blank image), it is forced to classify it into one of the two known categories. This can lead to misleading or incorrect predictions, as the model doesn't have an "irrelevant" class to output.

To address this, there are several fundamental approaches:
1.  **Retraining with an "Irrelevant" class**: The most robust solution is to retrain the model with an additional class for "irrelevant" images. This would turn the binary classifier into a multi-class classifier (Normal, OSCC, Irrelevant).
2.  **Out-of-Distribution (OOD) Detection**: Implement advanced techniques to detect if an input image falls outside the data distribution the model was trained on (i.e., biopsy images). If it's OOD, it can be flagged as irrelevant before being fed to the core classifier.
3.  **Heuristics**: Utilize simple, rule-based image processing techniques to identify clearly non-biopsy images based on low-level features. This is a preliminary approach to filter obvious cases.

### Proposing a Conceptual Heuristic for `is_irrelevant_image`

For a quick and conceptual solution, we can integrate a simple heuristic into the `is_irrelevant_image` function. Biopsy images typically exhibit intricate textures and varied color distributions due to their cellular and tissue structures. In contrast, many irrelevant images might be visually "simple," such as:

*   Images that are predominantly a solid color.
*   Images with very low detail or significant blur.
*   Completely blank or uniform images.

A straightforward heuristic could involve calculating the **standard deviation of pixel intensities** across the image. If the standard deviation is very low, it indicates that most pixels in the image have similar intensity values, suggesting a uniform or near-uniform image. Such an image is unlikely to be a detailed biopsy slide.

**Limitations of this Heuristic:**

*   **False Positives:** A genuine biopsy image that happens to be very uniform (e.g., a slide of a very homogenous tissue area) might be incorrectly flagged as irrelevant.
*   **False Negatives:** A complex irrelevant image (like a detailed photograph of something other than a biopsy) would likely have a high standard deviation and would therefore pass this check, still leading to a forced classification by the model.
*   **Threshold Sensitivity:** The choice of the standard deviation threshold is arbitrary and might require extensive tuning to be effective, and it might not generalize well across different types of irrelevant images.

Despite these limitations, this heuristic can serve as a quick initial filter for *obviously* irrelevant inputs, such as solid-color images or corrupted files that appear blank.

### Discussion and Need for Robust Solutions

While the implemented heuristic can catch obvious cases like solid-colored images, it's a superficial solution. It fails to address:
*   Complex non-biopsy images (e.g., photos of objects, text documents).
*   Subtle differences between true biopsy images and other medical images that are not biopsy-related.

For a production-ready system, more robust solutions are essential:

1.  **Retraining with an "Irrelevant" Class**: This is the recommended approach for improved accuracy and generalizability. By including diverse "irrelevant" images in the training dataset and assigning them to a dedicated class, the model learns to distinguish between relevant biopsy images and all other types of images.
2.  **Out-of-Distribution (OOD) Detection**: Techniques like using uncertainty quantification, reconstruction errors from autoencoders, or specialized OOD detection models (e.g., using Mahalanobis distance in feature space) can provide a more principled way to identify inputs that differ significantly from the training data distribution.

In conclusion, the simple heuristic helps to prevent gross misclassifications of visually uniform irrelevant images. However, to truly robustify the system against arbitrary irrelevant inputs, the model would need to be retrained with an explicit "Irrelevant" class or augmented with advanced Out-of-Distribution detection capabilities.

#### Explanation of the Irrelevant Image Problem

The current model is a binary classifier trained exclusively on 'Normal' and 'OSCC' biopsy images. When presented with an image that is not a biopsy (e.g., a hand, a landscape, or a different medical image type), the model is forced to categorize it into one of its known classes ('Normal' or 'OSCC'). This leads to misleading and incorrect predictions, as the model lacks the capacity to identify and reject irrelevant inputs, effectively misclassifying out-of-distribution data as one of its in-distribution classes.

#### Fundamental Approaches to Solve the Problem

There are three main approaches to address the issue of misclassifying irrelevant images:

1.  **Retraining with an Additional 'Irrelevant' Class:** This involves expanding the training dataset to include a new category for irrelevant images. The model would then be retrained as a multi-class classifier (e.g., 'Normal', 'OSCC', 'Irrelevant'). This allows the model to explicitly learn and classify inputs that do not belong to the primary diagnostic categories.

2.  **Implementing Out-of-Distribution (OOD) Detection:** OOD detection techniques aim to identify inputs that differ significantly from the data the model was trained on. Instead of classifying OOD data into one of the existing classes, these methods flag them as 'unknown' or 'out-of-distribution'. This approach keeps the original binary classifier intact for relevant images while adding a separate mechanism to handle irrelevant ones.

3.  **Utilizing Simple Heuristics based on Low-Level Image Features:** This involves defining rule-based criteria to pre-filter images that are clearly non-biopsy. For instance, checks could be implemented for image size, aspect ratio, color profiles, or texture characteristics that are typical of biopsy images versus general irrelevant images. This acts as a preliminary filter, preventing obviously irrelevant images from ever reaching the main classification model.

### Conceptual Heuristic: Standard Deviation of Pixel Intensities

1.  **Explanation of Heuristic**: A conceptual heuristic to flag irrelevant images could involve calculating the standard deviation of pixel intensities across an image. This can be applied to grayscale images, or individually to each color channel (Red, Green, Blue) and then averaged, to quantify the variation in brightness or color within the image.

2.  **Rationale for Choosing this Heuristic**: Biopsy slides, by their very nature, contain intricate cellular structures, varied staining patterns, and complex tissue textures. These visual characteristics lead to significant variations in pixel intensities across the image, resulting in a *high standard deviation*. Conversely, many irrelevant images (e.g., solid color backgrounds, plain text documents, heavily blurred images, or images with very uniform content) would exhibit much less variation in pixel intensities, leading to a *low standard deviation*. Therefore, images with a standard deviation below a certain empirical threshold could be flagged as potentially irrelevant.

3.  **Limitations of this Heuristic**:
    *   **False Positives**: A genuine biopsy image that happens to be visually uniform (e.g., a very clear and sparsely populated area of a slide, or a slide with a very homogeneous background) might have a low standard deviation and could be incorrectly flagged as irrelevant.
    *   **False Negatives**: A complex, but entirely irrelevant, image (e.g., a busy street scene, a highly textured abstract art piece) might have a high standard deviation, similar to a genuine biopsy slide, and would therefore pass the heuristic, leading to a false negative.
    *   **Threshold Sensitivity**: The effectiveness of this heuristic is highly dependent on the chosen threshold for the standard deviation. Setting a threshold too low could lead to many false positives, while setting it too high could result in many false negatives. This threshold would likely need extensive empirical tuning and might vary depending on the specific characteristics of the biopsy slides being analyzed.

````python
import os
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from PIL import Image

# Load best model
model = load_model("Oral_Cancer_Model.keras", compile=False)

class_names = ['Normal', 'OSCC','Irrelevant image']

# Recompile with binary_crossentropy as the original model used sigmoid output for binary classification
model.compile(optimizer='adamax',
              loss='binary_crossentropy',
              metrics=['accuracy'])

def is_irrelevant_image_heuristic(img_path, std_threshold=10.0):
    """Checks if an image is irrelevant based on pixel intensity standard deviation."""
    if not os.path.exists(img_path):
        print(f"Error: Image file not found at {img_path}")
        return False # Cannot determine, let the main predictor handle

    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print(f"Error: Unable to load image from {img_path} for irrelevance check.")
        return False

    # Calculate standard deviation of pixel intensities
    std_dev = np.std(img)
    # print(f"Image: {img_path}, Std Dev: {std_dev:.2f}") # For debugging

    if std_dev < std_threshold:
        print(f"Heuristic detected low pixel variance (Std Dev: {std_dev:.2f} < {std_threshold:.2f}). Marking as irrelevant.")
        return True
    return False

def predict_single_image(img_path, std_threshold=10.0, confidence_threshold=0.90):
    """Predicts the class of a single image, including an irrelevance check."""

    # First, check with the simple pixel variance heuristic
    if is_irrelevant_image_heuristic(img_path, std_threshold):
        return class_names[2], 1.0 # Return 'Irrelevant image' with high confidence

    if not os.path.exists(img_path):
        print(f"Error: Unable to load image from {img_path}")
        return None, None

    img = cv2.imread(img_path)
    if img is None:
        print(f"Error: Unable to load image from {img_path}")
        return None, None

    img = cv2.resize(img, (224, 224))   # Resize same as training
    img = img / 255.0                   # Normalize
    img = np.expand_dims(img, axis=0)   # Add batch dimension

    prediction = model.predict(img)
    predicted_prob = prediction[0][0]

    # Apply the confidence-based irrelevance heuristic
    if predicted_prob > confidence_threshold:
        predicted_class_idx = 1  # OSCC
        confidence = predicted_prob
    elif (1 - predicted_prob) > confidence_threshold:
        predicted_class_idx = 0  # Normal
        confidence = (1 - predicted_prob)
    else:
        # If neither class has high enough confidence, it's irrelevant
        print(f"Low confidence prediction (OSCC: {predicted_prob*100:.2f}%, Normal: {(1-predicted_prob)*100:.2f}%). Marking as irrelevant.")
        return class_names[2], 1.0 # Return 'Irrelevant image' with high confidence

    print(f"Prediction: {class_names[predicted_class_idx]} (Confidence: {confidence*100:.2f}%) ")
    return class_names[predicted_class_idx], confidence

print("Modified predict_single_image and is_irrelevant_image_heuristic functions defined.")
````

````python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
# from google.colab import files  <-- No longer needed for upload
import io
import os # <-- Added to check if file exists

# --- 1. Load the Keras Model ---
# NOTE: Make sure you have uploaded "oral_cancer_detection.keras" to your Colab session!
try:
    # You can upload your model using the file browser on the left in Colab
    model_path = "Oral_Cancer_Model.keras"

    custom_objects = {"softmax_v2": tf.nn.softmax}
    model = load_model(
        model_path,
        custom_objects=custom_objects,
        compile=False
    )

    # Re-compile the model with BINARY loss
    model.compile(
        optimizer='adamax',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    print("✅ Model loaded successfully!")

except Exception as e:
    print(f"❌ Error loading model: {e}")
    print("---")
    print("👉 PLEASE MAKE SURE you have uploaded your 'oral_cancer_detection.keras' file to Colab.")
    model = None

# --- 2. Define Constants and Heuristic Function ---
# These are identical to your Flask app
class_names = ['Normal', 'OSCC', 'Irrelevant image']
STD_THRESHOLD = 10.0
CONFIDENCE_THRESHOLD = 0.90
IMG_SIZE = (224, 224) # Added for resizing

def is_irrelevant_image_heuristic(img_bytes, std_threshold=STD_THRESHOLD):
    """
    Checks if an image is likely irrelevant (e.g., solid color, blank)
    by checking the standard deviation of its grayscale pixel values.
    """
    try:
        nparr = np.frombuffer(img_bytes, np.uint8)
        img = cv2.imdecode(nparr, cv2.IMREAD_GRAYSCALE)
        if img is None:
            return False

        std_dev = np.std(img)
        return std_dev < std_threshold
    except Exception as e:
        print(f"Heuristic check failed: {e}")
        return False

# --- 3. Define the Main Prediction Function ---
def predict_image(img_bytes):
    if model is None:
        print("Model is not loaded. Cannot predict.")
        return

    try:
        # --- Step 5a: Run Heuristic Check ---
        if is_irrelevant_image_heuristic(img_bytes):
            print("--- Prediction Result ---")
            print("Prediction: Irrelevant Image")
            print("Confidence: 100%")
            print("Reason: Image appears to be blank or solid color (low pixel variance).")
            return

        # --- Step 5b: Decode and Preprocess Image ---
        npimg = np.frombuffer(img_bytes, np.uint8)
        img = cv2.imdecode(npimg, cv2.IMREAD_COLOR)

        if img is None:
            print("Error: Failed to decode image.")
            return

        img_resized = cv2.resize(img, IMG_SIZE)
        img_normalized = img_resized / 255.0
        img_batch = np.expand_dims(img_normalized, axis=0)

        # --- Step 5c: Make Prediction ---
        prediction = model.predict(img_batch)
        predicted_prob = prediction[0][0]

        # --- Step 5d: Run Confidence Check ---
        print("--- Prediction Result ---")
        if predicted_prob > CONFIDENCE_THRESHOLD:
            prediction_label = "High Risk (OSCC Detected)"
            confidence = float(predicted_prob)
            print(f"Prediction: {prediction_label}")
            print(f"Confidence: {confidence*100:.2f}%")

        elif (1 - predicted_prob) > CONFIDENCE_THRESHOLD:
            prediction_label = "Low Risk (Normal)"
            confidence = float(1 - predicted_prob)
            print(f"Prediction: {prediction_label}")
            print(f"Confidence: {confidence*100:.2f}%")

        else:
            prediction_label = "Uncertain / Low Confidence"
            confidence = float(max(predicted_prob, 1 - predicted_prob))
            print(f"Prediction: {prediction_label}")
            print(f"Confidence: {confidence*100:.2f}%")
            print("Reason: The model is not confident. Image may be blurry, out of focus, or an invalid slide.")

    except Exception as e:
        print(f"Error during prediction: {e}")

# --- 4. Run the Prediction on a Local File ---
if model is not None:
    # -----------------------------------------------------------------
    # 👇 EDIT THIS LINE to point to your test image in Colab
    TEST_IMAGE_PATH = "Aboutpage2.jpg"
    # e.g., "my_test_photo.png" or "/content/sample_data/mnist_test.png"
    # -----------------------------------------------------------------

    print(f"Attempting to predict file: {TEST_IMAGE_PATH}")

    if not os.path.exists(TEST_IMAGE_PATH):
        print(f"❌ ERROR: File not found at path: {TEST_IMAGE_PATH}")
        print("Please make sure the file is uploaded to Colab and the path is correct.")
    else:
        try:
            # Read the file bytes
            with open(TEST_IMAGE_PATH, "rb") as f:
                file_bytes = f.read()

            print(f"\nProcessing file: {TEST_IMAGE_PATH}")
            # Run the prediction
            predict_image(file_bytes)

        except Exception as e:
            print(f"Error reading or processing file: {e}")
````
